# Preprocessing

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Price',axis=1)
```

```python
X
```

```python
y = df['Price']
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
X_train.columns
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer,make_column_transformer
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
cT1 = make_column_transformer((OneHotEncoder(categories=ohe.categories_,drop='first'),[0,1,4]),
                              (StandardScaler(),[2,3])
                                   ,remainder='passthrough')
```

```python
#cT2 = make_column_transformer((StandardScaler()),[],remainder='passthrough')
```

# Model

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
```

```python
pipe = make_pipeline(cT1,linear)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
#checking corresponing to which random state we get best r2_score,for 1000 iterations of random states
r_score = []
for i in range(0,1001):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=i)
    cT1 = make_column_transformer((OneHotEncoder(categories=ohe.categories_,drop='first'),['name','company','fuel_type']),
                              (StandardScaler(),['year','kms_driven'])
                                   ,remainder='passthrough')
    pipe = make_pipeline(cT1,linear)
    pipe.fit(X_train,y_train)
    y_pred = pipe.predict(X_test)
    r_score.append(r2_score(y_test,y_pred))
```

```python
max(r_score)
```

```python
r_score.index(max(r_score))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=516)
cT1 = make_column_transformer((OneHotEncoder(categories=ohe.categories_,drop='first'),[0,1,4]),
                              (StandardScaler(),[2,3])
                                   ,remainder='passthrough')
pipe = make_pipeline(cT1,linear)
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
r2_score(y_test,y_pred)
```
